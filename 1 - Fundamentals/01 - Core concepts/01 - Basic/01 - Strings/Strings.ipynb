# 1 - String Format

## end
END command will put the down line in front of the up line, everything inside it will be used instead of a paragraph.
- `end=' >>> ' will add 3 right arrows instead of paragraph`

## formatting numbers in variables
-  `{:.3f}` To consider 3 variables after comma.
- ` {:-^20} `Add 20 - and center the variable
-  `\n` to break the line


## .format
Will fill the specific location marked as {} with the variable in left to right order
-` print('{}{}).format(variable1,variable2...)`

```python
n1 = 5
n2 = 3
m = n1/n2
name = 'Jose'

print('\nThe division: {:.3f}' .format(m), end=' >>> ')
print('{} / {}'.format(n1, n2))

#USING FORMAT
print('Nice to meet you {:-^20}'.format(name)) #one example using .format
```

## f strings
Similar to .format, but inside the {} we have to fill the variable. Just need to add f before the '
- `print(f'{variable1}...')`

```python
age = 33
salary = 1256.365484

print(f'{name} is {age} and his salary is {salary:.2f}\n') #.2f to consider only 2 decimals after the comma
print(f'{name:-^20}') #Centered in the middle of 20 ^
print(f'{name:20}') #20 invisible spaces in front
print(f'{name:^20}') #20 invisible spaces but centered
print(f'{name:-<20}') #20 instead of spaces hifen
print(f'{name:->20}') #20 instead of spaces hifen


```

# 2 - Strings Management

## .len()
Counts the size of characters of the variable, considering even the invisible spaces.







## .split()


Divide the variable phrase into a list where spaces determine each element<br>
To show elements from lists in front of the variable we add [0,1,2...] <br>
decide the which element of the list to see as 0 is the first term of the list<br>
<br>
ps: LISTS are collective elements to learn more go to COLLECTIONS.

```python
phrase = '01 - Python as a programming Language  '
divided = phrase.split()

print(f'Normal: {phrase} it has {len(phrase)} characters')
```

## .join()


Join the list, in the case bellow using - as space.

```python
join = '-'.join(divided)
print(f'Join (-): {join} it has {len(join)} characters')
```

## .strip()



Cuts spaces on the beginning and ending of variables

```python
strip = phrase.strip()
print(f'Strip: {strip} it has {len(strip)} characters')
```

## .upper() and .lower()


Set letters to CAPITALIZE, LOWERCASE<br>
    `.upper()[0]` In this case I am making the first letter as capital

```python
capitalize = phrase.upper()
print(f'Capitalize: {capitalize}')
```

## .count('x')


Count the number of characters of the variable, considering even the invisible spaces. _**(Case sensitive)_**<br>
It is good to work this with `.lower` or `.upper`

```python
print(f'Counting efficient "o": ',phrase.upper().count('O'))
```

## .replace('old','new')

```python
replace = phrase.replace('01 - Python', 'Android')
print(f'Replace: {replace} it has {len(replace)} characters')
```

## .find('word')


It will tell exactly where it will find the word first time. If returns -1 this means it could not find the word

It is good to work this with .lower or .upper<br>
To make the command works right to left we just add an 'r' in front of the command<br>

  Examples:<br>
 ` print('Test {}'.format(variable.find('A')+1))`<br>
  `print('Test {}'.format(variable.rfind('A')+1))`

```python
print('The word language appears in ', phrase.lower().find('language'))
```

## Slicing

Strings are considered as a list of letters' with that in ind we can work SLICING techniques for STRINGS<br>
<br>
`phrase[start:end:step]:`<br>
`phrase[3]` Get only the third element in this case 'T'<br>
`phrase[3:13]` Get the 3rd element until the 13th (not considering 13th)<br>
`phrase[::2]` Without start. Without ending, going 2 by 2<br>

```python
print(phrase[6])
print(phrase[3:13])
print(phrase[::2])
```

## Strings as List

### Strings Immutable
Strings are immutable so to use the functions bellow it is important to use another variable if we have plans to use
the variable. If the plan is just to print the variable it can be done inside, but the variable will not change.





```python
a = 'Hello World'
print('Here is UPPER: ',a.upper())
print('Here the real value not changed of a: ', a)
b = a.upper()
print('This is the real vale of b: ',b)
```

### Stings works like LISTS
'Strings are considered as a list of letters' with that in ind we can work SLICING techniques for STRINGS

```python
print(f'As a LIST: {divided}')
print(f'As a LIST by element (first element): {divided[0]}')
print(f'As a LIST by element (second element): {divided[1]}')
```
