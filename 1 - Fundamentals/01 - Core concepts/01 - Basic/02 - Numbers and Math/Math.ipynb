```python
n = 2
m = 5

o = m**n
o1 = pow(m,n) #another 5.1 - For potency
p = m//n
q = m%n

print(f"""
As base {m} and potency of {n} we have {o} and {o1}
The complete division between then is {p}
The rest is {q}
""")
```
