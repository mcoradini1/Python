## Single Variables:<br>


Do not create a bond within the variables<br>
`x = 3 `  (create first object 3, then create first variable x and insert the reference for the object)<br>
`y = x  ` (object x already exists, create a new variable y and then reference an object)<br>
`y = y - 1` (remove the reference from 3, to 2...)<br>

```python
x = 3
y = x
y = y - 1

print(x,y)
```

# List, Dicts:

`l1 = [1,2,3] `(create the first object [1,2,3] list and assign it to L1)<br>
`l2 = l1` (assign the object to this variable creating a bond with then)<br>
`l1[0] = 24 `(both l1 and l2 are equal...)<br>
`l1 == l2 ` (true)<br>
`l1 is l2` (true)<br>

```python
m1 = [1,2,3,4,5]
m2 = [1,2,3,4,5]

print(m1 == m2)

#these two are similar:
print(m1 is m2)
print(id(m1) == id(m2))
```

# Dissecting an Variable

`variable.dissecting-function()`<br>
There are other examples such as isprint, isspace...

## isnumeric()

```python
n = input('Type a number: ')
print(n.isnumeric(), '5.1 - For numbers\n')

#This check if it is a number (answer in boolean)
```

## isalpha()

```python
m = input('Type some letters: ')
print(m.isalpha(), '5.1 - For letters\n')
#This check if it is alphabetic (answer in boolean)
```

## isalnum()

```python
o = input('Type numbers and letters: ')
print(o.isalnum(), '5.1 - For numbers and letters\n')
#This check if it is alphanumeric (answer in boolean)
```

## isupper()

```python
p = input('Type letters in capitalcase: ')
print(p.isupper(), '5.1 - For all letters in CAPITAL\n')
#This check if it is CAPITAL letters (answer in boolean)
```

## islower()

```python
q = input('Type lowercase letters: ')
print(q.islower(), '5.1 - For lowercase letters\n')

```
