```python
import matplotlib.pyplot as plt
import pandas as pd

whisky = pd.read_csv('whiskies.txt')
whisky['Region'] = pd.read_csv('regions.txt')
whisky.head() # first 5
whisky.tail() # last 5
```

```python
whisky.iloc[0:10] #specifying rows from 0 to 10 (not considering 10)
whisky.iloc[0:10, 0:5] #specifying rows from 0 to 10 (not considering 10) and columns (0 to 5)
whisky.columns
flavors = whisky.iloc[:, 2:14] #take all lines from columns 2 until 14 (flavors)
flavors
```

```python

#correltion between flavors
#the function used for correlation will be pearson correlation
#1 is higher...

corr_flavors = pd.DataFrame.corr(flavors)

plt.figure(figsize=(10,10))
plt.pcolor(corr_flavors)  #How can you plot a correlation matrix by color?
plt.colorbar() #What is the matplotlib.pyplot function that plots a colorbar on the side of a plot?
plt.show()

#correlation between the same is 1 (diagonal line)
#but there are someother strong correlations like heavybody and smokiness as floral is opposite of  medicinal note or fullbody
#we can see the correlation amoung whiskies...

corr_whiskies = pd.DataFrame.corr(flavors. transpose())
plt.figure(figsize=(10,10))
plt.pcolor(corr_whiskies)  #How can you plot a correlation matrix by color?
plt.axis('tight') #to show just the 86 types only...
plt.colorbar() # What is the matplotlib.pyplot function that plots a colorbar on the side of a plot?
plt.show()

```

```python
from sklearn.cluster import SpectralCoclustering
import numpy as np

#Clustering Whiskies by flavor profile
#we will do that using a clustering method from scikit-learn machine learning
#spectral co-clustering
#eigenvaluer and eigenvectors (more in matrix or linear algebra)

model = SpectralCoclustering(n_clusters=6, random_state=0).fit(corr_whiskies)  #A method for finding clusters of objects by the similarity of their attributes
#each row in this array identifies a cluster ranging from 0 to 5. and each column identify a row in correlation matrix here ranging from 0 to 85

#model.rows_
np.sum(model.rows_, axis=1) #0 rows, 1 columns
#19 whiskies from cluster 0, 5 whiskeys from clustaer 1...)
np.sum(model.rows_, axis=0) #only 1 as we know
model.row_labels_ #this shows each element inside the array and (must be from 0 to 5) because we specify 6 clusters...
```
