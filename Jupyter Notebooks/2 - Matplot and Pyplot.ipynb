### Plot Customization
`linewidth = value`
`markersize = value`
`label = name to use for legend  `

#### Blue circles (linewidth and marker size edit)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 20)  #create a 1d array
y = x ** 2
plt.plot(x, y, 'bo', linewidth=2, markersize = 4)
```

#### Green squares + line (linewidth and marker size edit)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 20)  #create a 1d array
y = x ** 2
plt.plot(x,y, 'gs-', linewidth = 2, markersize = 4)
```

#### Red Diamonds + line (linewidth and marker size edit)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 20)  #create a 1d array
y = x ** 2
plt.plot(x,y, 'rd-', linewidth = 2, markersize = 10)
```

#### Axis naming and positioning
`plt.xlabel($x$)` - $ give a specific style, not necessary
`plt.ylabel(y)`
`plt.axis([xmin, xmax, ymin, ymax])`

Plotting two lines in the same image.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,20)
y = x ** 2
y2 = x ** 1.5

plt.plot(x,y, 'bo-', linewidth = 2, markersize = 4)
plt.plot(x,y2, 'gs-', linewidth = 2, markersize = 4)
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.axis([-5, 10.5, -5, 105])
```

#### Labelling and Saving
`plt.legend(loc='upper left')`
`plt.savefig('____.jpg')`

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,20)
y = x ** 2
y2 = x ** 1.5

plt.plot(x,y, 'bo-', linewidth = 2, markersize = 4, label = 'First')
plt.plot(x,y2, 'gs-', linewidth = 2, markersize = 4, label = 'Second')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
#plt.savefig('_________.pdf')  #here I can decide the format that it will save. (pdf, png, img, jpg)

```

### Logarithmic Plot Customization
useful when i have parabolas to make it linear.
`plt.semilogx()` for x
`plt.semilogy()` for y
`plt.loglog()` for both


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.logspace(-1,1, 40)
y = x ** 2
x2 = x ** 1.5
y2 = x2 ** 4

plt.loglog(x,y, 'bo-', linewidth = 2, markersize = 4, label = 'x1')
plt.loglog(x2,y2, 'gs-', linewidth = 2, markersize = 4, label = 'y1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.show()
```

## Histogram
`plt.hist(x)`

### Default Histogram (with 10 bins)
bins is the space of each graph
`np.random.normal(size=1000)`: this follow creating a normal distribution with 1000 values inside.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(size=1000)
plt.hist(x)
plt.show()  #not necessary
```

### Modified Histogram
`plt.hist(x, density=True, bins = np.linspace(-5,5,21))`
`density` = Instead of number of observation, it shows proportion (old  normed)
`bins` = the amount of 'columns' in my graph

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(size=1000)
plt.hist(x, density=True, bins = np.linspace(-5,5,21))
plt.show()  #not necessary
```

```python
a = np.random.gamma(2,3,100)
plt.figure()
plt.subplot(2,2,1)
plt.hist(a, bins = 30, density = True, cumulative = True, histtype= 'step')
plt.subplot(2,2,2)
plt.hist(a, bins = 30, density = True)
plt.subplot(223)
plt.hist(a, bins = 15, cumulative = True, histtype= 'step')
plt.subplot(224)
plt.hist(a, bins = 15, cumulative = True, histtype= 'stepfilled')
plt.show()
```
