### Roll a die x times
`for k in range (100):`
`x = random.choice([1, 2, 3, 4, 5, 6])`

The probability of each number is the same, that brings the theory that the histogram tends to be FLAT, as we increase the number of visualizations.
 To check that see the next two examples.

#### Roll a die 100 times

```python
import random
import matplotlib.pyplot as plt
import numpy as np

rolls = []
for k in range (10):
    x = random.choice([1, 2, 3, 4, 5, 6])
    rolls.append(x)
print(len(rolls))
print(rolls)

plt.hist(rolls, bins = np.linspace(0.5,6.5,7))
plt.show()
```

#### Roll a die 10000 times

```python
import random
import matplotlib.pyplot as plt
import numpy as np
rolls = []

for k in range (10000):
    x = random.choice([1, 2, 3, 4, 5, 6])
    rolls.append(x)
print(len(rolls))
print(rolls)

plt.hist(rolls, bins = np.linspace(0.5,6.5,7))
plt.show()
```

### Roll x Dice x Times
for rep in range (100):

    for k in range (10):



Central Limit Theorem (CLT) says about the sum or larger numbers of random variables, regardless of their distribution it will be proximate to a normal distribution.

```python
import random
import matplotlib.pyplot as plt

y_sum = []
x = []
for rep in range (10000):
    y = 0
    for k in range (10):
        x = random.choice([1, 2, 3, 4, 5, 6])
        y = y + x
    y_sum.append(y)

print(len(y_sum),max(y_sum),min(y_sum))
plt.hist(y_sum)
plt.show()
```

## Python Time Library
`time.time()`
`time.perf_counter()`
`time.process_time()`


#`time.time()` , `time.perf_counter()` , `time.process_time()`  = Similar functions

```python
import time
import random

start = time.time()     #check the time that start

ys = []
x = []
for rep in range (1000000):
    y = 0
    for k in range (10):
        x = random.choice([1, 2, 3, 4, 5, 6])
        y = y + x
    ys.append(y)

print(len(ys),max(ys),min(ys))


end = time.time()        #check time that end

final = end - start      #to be able to know the process time
print(final)
```
