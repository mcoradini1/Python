## 2. np.array()
**shape** and **size** are data attribute of arrays, they do not need () like methods.

### Vector (1 Dimension)

```python
import numpy as np
x = np.array([1,2,3])
y = np.array([2,4,6])
print(x)
print(y)

print(f'{x.shape}')
print(f'{x.size}')
```

### Matrix (2 Dimension)

```python
import numpy as np
z = np.array([[1,3], [5,9]])
print(z)
```

### Transpose Matrices

```python
import numpy as np
z = np.array([[1,3], [5,9]])
z1 = z.transpose()
print(z1)
```

### Slicing

#### Vectors (1 dimension)

```python
import numpy as np
x = np.array([1,2,3])
y = np.array([2,4,6])

print(x[0])
print(x[0:2])

z = x + y

print(f'{x} + {y} = {z}')
```

#### Matrices (2 dimensions+)

##### Columns

```python
import numpy as np
x2 = np.array([[1,2,3], [4,5,6]])
y2 = np.array([[2,4,6], [8,10,12]])

print(x2[:,0])                #get all numbers from the first column
print(y2[:,0])                # get all number from the first column
print(x2[:,0] + y2[:,0])      #sum term by term of the specified columns
```

##### Rows

```python
import numpy as np
x2 = np.array([[1,2,3], [4,5,6]])
y2 = np.array([[2,4,6], [8,10,12]])
print(x2)
print('\n')

print('test 1',x2[:2])
print('\n')
print(x2[1,:])
print('\n')
print(x2[0])          #same as before just demonstrating (less typing)
print('\n')
print(y2[0,:])
print('\n')
print(x2[0,:] + y2[1,:])
```

## 3. Indexing
Can use arrays or lists to do that. It is important to use it to don't create bond between...

```python
import numpy as np
index_array = np.array([0,2,3])
index_list = [0,2,3]

z = np.array([1,3,5,7,9])
z1 = z + 1

print('1: ',z1[index_array]) #Both will give me the term 0,2,3 of z1
print('2: ',z1[index_list])

print('3: ',z1 > 6)    #Show True or False each term

print('4: ',z1[z1>6])

z3 = z1 > 6
z4 = np.array([8,16,32,33,0])
print('5: ',z3)

#When z3 is created it will be a boolean array, the examples bellow demonstrate that it will always show terms 3 and 4 (no matter the value there)
print('6: ',z[z3])
print('7: ',z1[z3])
print('8: ',z4[z3])
```

### Copying
To work with arrays it's important to note that different from lists slicing will create a bond between them, to sort it out we use INDEXING.

#### incorrectly (bonding them)

```python
import numpy as np

z = np.array([1,3,5,7,9])

print(f'This is z1: {z}')

w = z[0:3]
print(f'this is w: {w}')

w[0] = 600

print(f'{z} and are bonded {w}')
```

#### correctly (not bonding)

```python
import numpy as np
z = np.array([1,3,5,7,9])

index_z = np.array([0,1,2])
w = z[index_z]
w[0] = 600

print(f'{z} and are bonded {w}')
```

## 4. Building linear arrays
to create an array linearly spaced

### np.linspace()
`np.linspace(start,stop,how many terms)`

```python
import numpy as np
x = np.linspace(0,100,10)
print(x)
```

### np.logspace()
`nplogspace(start,stop,how many terms)`

```python
import numpy as np
y = np.logspace(1,2,10) #log 10 = 1, log 100 = 2 and 10 terms
print(y)
```

#### Specify logbase

```python
import numpy as np
print(np.logspace(np.log10(250),np.log10(500),10))
```

## 6. Random
`np.random.random(x)`
`np.random.normal(a,b,x))`
`np.random.randint(low, high, x)`
a = mean = 0 (usual)
b = standard deviation = 1 (normal distribution)
x = array (for more than 1d it's necessary to use x as a tuple)

### Question
What does numpy.random.random((5,2,3)) do?

### Vector

```python
import numpy as np
x = np.random.random(5)
y = np.random.normal(0,1,5)
z = np.random.randint(1,7, 5)
print(x)
print(y)
print(z)
```

### Matrix

```python
import numpy as np

x = np.random.random((2,5))
y = np.random.normal(0,1, (2,5))
z = np.random.randint(1,7, (2,5))

print(x)
print('\n')
print(y)
print('\n')
print(z)
```

#### Some Functions
`np.sum(array, axis = x)`
x = 0 (first term) = Rows
x = 1 (second term) = Columns
x = 2+ (cases of arrays with more dimensions)

```python
x = np.random.randint(1,7, (3,10))
print(x)
print('array shape:', x.shape)

print('array sum:',np.sum(x))
print('rows sum:',np.sum(x, axis=0)) # sum of all rows of the array
print('columns sum:',np.sum(x, axis=1)) #2 sum of all columns array
```
