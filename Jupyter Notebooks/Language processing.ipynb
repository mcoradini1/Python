```python
text = 'This is my test text. We are keeping this text short to keep things manageable'

def count_words_version1(txt):
    """
    Counts the amount of times each word occurs in text (str). Return a dictionary with every unique word count.
    :param txt:
    :return:
    """
    words_count = {}
    for word in txt.split(' '):
        if word in words_count:   #known word
            words_count[word] += 1
        else:                           #unknown word
            words_count[word] = 1
    return words_count

x = count_words_version1(text)

```

## Word Statistics Function

```python
def read_book(book_title):
    """
    Read a book and return it as a string.
    :param book_title:
    :return:
    """
    with open(book_title, 'r', encoding="utf-8") as current_file:
        book_text = current_file.read()
        book_text = book_text.replace('\n', '').replace('\r', '')

    return book_text

def count_words_version_3(text_count):
    """
    Counts the amount of times each word occurs in text (str). Return a dictionary with every unique word count. Skip Punctuation.
    :param text_count:
    :return:
    """

    text_count  = text_count.lower()
    skips = ['.' , ',' , ';' , "'" ,'"',]
    for char in skips:
        text_count = text_count.replace(char, '')

    words_count = Counter(text_count.split(' '))

    return words_count

def word_statistics(stats):
    """
    Return number of unique words and wpd frequency
    :param stats:
    :return:
    """
    word_unique = len(stats)
    word_count = stats.values() #the variable is a dictionary
    return word_count, word_unique

#CREATING A VARIABLE WITH THE BOOK STRING INSIDE
book_english = (read_book(r'C:/Programacao/Python/Jupyter Notebooks/python_case_studies/language_processing/English/shakespeare/Romeo and Juliet.txt'))
book_german = (read_book(r'C:/Programacao/Python/Jupyter Notebooks/python_case_studies/language_processing/German/shakespeare/Romeo und Julia.txt'))

#COUNTING THE WORDS IN EACH
count_english = count_words_version_3(book_english)
count_german = count_words_version_3(book_german)

#EACH STATISTIC
stats_english = word_statistics(count_english)
stats_german = word_statistics(count_german)

#STORAGING THE TUPLE INFORMATION INTO 2 DIFFERENT VARIABLES
(english_count, english_unique) = stats_english
(german_count, german_unique) = stats_german

#PRINTING RESULTS
print(f'English total words: {sum(english_count)}| Total unique words: {english_unique}')
print(f'German total words: {sum(german_count)}| Total unique words: {german_unique}')
print(f'The difference of total word is {sum(english_count) - sum(german_count)} and the unique words is {-1 * (english_unique - german_unique)}')




```

```python
def read_book(book_title):
    """
    Read a book and return it as a string.
    :param book_title:
    :return:
    """
    with open(book_title, 'r', encoding="utf-8") as current_file:
        book_text = current_file.read()
        book_text = book_text.replace('\n', '').replace('\r', '')

    return book_text

def count_words_version_3(text_count):
    """
    Counts the amount of times each word occurs in text (str). Return a dictionary with every unique word count. Skip Punctuation.
    :param text_count:
    :return:
    """

    text_count  = text_count.lower()
    skips = ['.' , ',' , ';' , "'" ,'"',]
    for char in skips:
        text_count = text_count.replace(char, '')

    words_count = Counter(text_count.split(' '))

    return words_count

def word_statistics(stats):
    """
    Return number of unique words and wpd frequency
    :param stats:
    :return:
    """
    word_unique = len(stats)
    word_count = stats.values() #the variable is a dictionary
    return word_count, word_unique

#-----------------------------------------------------------------------------------------------------------
import os
import pandas as pd
book_directory = './python_case_studies/language_processing'
stats = pd.DataFrame(columns = ('language', 'author', 'title','length','unique'))
title_num = 1


os.listdir(book_directory) #check what is in that path

for language in os.listdir(book_directory):

    for author in os.listdir(os.path.join(book_directory + '/' + language)):

        for title in os.listdir(os.path.join(book_directory + '/' + language + '/' + author )):
            input_file = book_directory + '/' + language + '/' + author + '/' + title

            text = read_book(input_file)
            text1 = count_words_version_3(text)
            text2 = word_statistics(text1)
            (english_count, english_unique) = text2

            stats.loc[title_num] = language, author.capitalize(), title.replace('.txt', ''), sum(english_count), english_unique #insert a row to a dataframe
            title_num += 1

print(stats)
stats.head() #five first
stats.tail() # five last

```
