### Function Sequence
`with`

```python
def read_sequence(input_file):
    """

    Reads and returns the input sequence with special characters removed.
    :param input_file: determine the document.text to be used
    :return: the sequence with special characters removed

    """

    #With command: It opens a file and uses it for the subsequent block of code only, and then closes the file.
    with open(input_file, 'r') as f:
        sequence = f.read()
        sequence = sequence.replace('\n','')
        sequence = sequence.replace('\r','')
        return sequence

```

### Function Translate

```python
def translate_codon(sequence):
    """

    :param sequence: add the DNA sequence to be translated
    :return: Sequence translated
    """
    table = {
    'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M',
    'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T',
    'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K',
    'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',
    'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L',
    'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P',
    'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q',
    'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R',
    'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V',
    'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A',
    'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E',
    'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G',
    'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S',
    'TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L',
    'TAC':'Y', 'TAT':'Y', 'TAA':'_', 'TAG':'_',
    'TGC':'C', 'TGT':'C', 'TGA':'_', 'TGG':'W',
    }


    protein = ''
    if len(sequence) % 3 == 0:
        for i in range(0,len(sequence),3):
            codon = sequence[i : i+3]
            protein += table[codon]
    return protein



```
