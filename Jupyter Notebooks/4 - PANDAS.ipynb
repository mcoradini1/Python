## Series
`pd.Series([x1...], index = [' '...])`<br>
`____.index`<br>
`sorted.(____index)`<br>
`____.reindex(sorted(____.index))`<br>

**INDEX NOT SPECIFIED (STANDARD 1,2,3...)**

```python
import pandas as pd
series = pd.Series([6,3,8,6])
series
```

**INDEX SPECIFIED**

```python
series = pd.Series([6,3,8,6], index = ['Maria','Emilia','Susana','Mario'])
series
```

**GATHER SPECIFIC INFORMATION**

```python
series['Maria']
```

```python
series[['Emilia','Mario', 'Maria']]
```

**SHOW INDEX**

```python
series.index
```

**INDEX ALPHABETIC ORDER**

```python
sorted(series.index)
```

**REINDEX: REORDER THE INDEXES WITH THEIR SPECIFIC VALUES**

```python
series.reindex(sorted(series.index))
```

**DICTIONARY AS BASE**

```python
age = {'Tim':29, 'Jim':44, 'Pam':19, 'Sam': 6}
series_from_dictionary = pd.Series(age)
print(series_from_dictionary)
```

## DataFrame
`pd.DataFrame(dictionary, the columns)`<br>
`pd.DataFrame(columns = (1,2,3... ))`<br>
`____.loc[x]` Shows specific row <br>
`____.columns` Shows only the columns (Index) <br>

**DATA FRAME CREATING AND ADDING**

```python
import pandas as pd

dataframe = pd.DataFrame(columns = ('name','age'))

dataframe.loc[1] = 'Marlon', 31
dataframe.loc[2] = 'Susana', 29
dataframe.loc[3] = 'Maria', 10
dataframe.loc[4] = 'Emmeline', 6
dataframe.loc[5] = 'Emilia', 45


print(dataframe.columns)
print('-'*35)
print(dataframe)
```

**DATA FRAME FROM DICTIONARY WITH LISTS<br>**
`pd.DataFrame(dictionary, the columns)`

```python
data = {'name':['Tim','Jim','Pam','Sam'],
        'age':[29, 44, 19,6],
    'address':['OX3012','OX1517','LO1511','OX1517']}


dataframe_from_dic = pd.DataFrame(data, columns=['name','age', 'address'])
print(dataframe_from_dic)
print('-'*35)
print(dataframe_from_dic.name)
```
